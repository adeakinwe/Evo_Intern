```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

## Storing the website's URL

```python
website = "https://ackodrive.com/cars/"
```

## Get Requests

```python
response = requests.get(website)
```

## Status Code

```python
response.status_code
```

## Soup Object

```python
soup = BeautifulSoup(response.content, 'html.parser')
```

```python
print(soup)
```

## Target Necessary Data

#### Title Of the Car

#### Fuel Type

#### Transmission

#### Price
